```python
import urllib.request
import re
import os

class MyFile:
    def init(self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.html = ""
    
    def read(self):
        if self.mode == "read":
            file = open(self.filename, "r")
            result = file.read()
            file.close()
            return result
        return ""
    
    def write(self, text):
        if self.mode == "write":
            file = open(self.filename, "w")
            file.write(text)
            file.close()
        elif self.mode == "append":
            file = open(self.filename, "a")
            file.write(text)
            file.close()
    
    def read_url(self):
        if self.mode == "url":
            response = urllib.request.urlopen(self.filename)
            data = response.read()
            self.html = data.decode("utf-8")
            return self.html
        return ""
    
    def count_urls(self):
        if self.mode == "url":
            if self.html == "":
                self.read_url()
            links1 = re.findall('href="https?://.*?"', self.html)
            links2 = re.findall("href='https?://.*?'", self.html)
            
            return len(links1) + len(links2)
        return 0
    
    def write_url(self, output_file):
        if self.mode == "url":
            if self.html == "":
                self.read_url()
            file = open(output_file, "w")
            file.write(self.html)
            file.close()
```
